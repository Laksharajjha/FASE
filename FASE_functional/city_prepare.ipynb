```python
import pandas as pd

cities = pd.read_csv('world_cities.csv')

cities.head()
```

```python
cities['name'].describe()
```

```python
cities = cities.drop_duplicates(subset='name',keep='first')
```

```python
cities['name'].describe()
```

```python
only_cities = list(cities['name'])
```

```python
only_cities_lower = []
for city in only_cities:
    only_cities_lower.append(city.casefold())

len(only_cities_lower)
```

```python
import json

with open('cities.json','w') as outfile:
    json.dump(only_cities_lower,outfile)
```
